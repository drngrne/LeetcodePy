```python
class Solution(object):
    def checkInclusion(self, s1, s2):
        len_s1 = len(s1)
        len_s2 = len(s2)

        hm_s1 = {}
        hm_s2 = {}

        for s in s1:
            if s in hm_s1:
                hm_s1[s] = hm_s1[s]+1
            else:
                hm_s1[s] = 1

        l=0
        r = len_s1 - 1

        while l<=r and r<len_s2:

            for z in s2[l : r + 1]:
                if z in hm_s2:
                    hm_s2[z] = hm_s2[z]+1
                else:
                    hm_s2[z] = 1

            if hm_s1 == hm_s2:
                return True

            l = l + 1
            r = r + 1
            hm_s2 = {}

        return False
```

```python
sola = Solution()
print(sola.checkInclusion('adc', 'dcda'))
```
