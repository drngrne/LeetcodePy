```python
class Solution(object):
    def reportSpam(self, message, bannedWords):
        hm = {}

        for m in message:
            hm[m] = hm.get(m, 0) + 1

        count = 0

        bannedWords = set(bannedWords)

        for bannedWord in bannedWords:
            if bannedWord in hm:
                count += hm[bannedWord]
                if count >= 2:
                    return True


        return False
```
