```python
class Solution(object):
    def trap(self, height):
        mstack = []
        water = 0

        for i, h in enumerate(height):
            while mstack and h > height[mstack[-1]]:
                bottom = mstack.pop()
                if not mstack:
                    break
                left = mstack[-1]
                bounded_height = min(height[left], h) - height[bottom]

                width = i - left - 1
                water += bounded_height * width
            mstack.append(i)

        return water



```

```python
sola = Solution()
print(sola.trap([1,3,2,1,2,1,5,3,2,1,2,1]))
```
