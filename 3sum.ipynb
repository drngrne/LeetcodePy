```python
class Solution(object):
    def threeSum(self, nums):
        result = []
        nums.sort()
        head = 0

        while head<=len(nums)-3:
            if (head>0) and (nums[head] == nums[head-1]):
                head=head+1
                continue


            l = head+1
            r = len(nums)-1

            while l<r:
                if nums[head]+nums[l]+nums[r] == 0:
                    result.append([nums[head], nums[l], nums[r]])
                    l=l+1

                    while nums[l] == nums[l-1] and l<r:
                        l=l+1

                elif nums[head]+nums[l]+nums[r] < 0:
                    l=l+1
                else:
                    r=r-1


            head=head+1

        return result
```

```python
sola = Solution()
print(sola.threeSum([2,-3,0,-2,-5,-5,-4,1,2,-2,2,0,2,-4,5,5,-10]))
```
