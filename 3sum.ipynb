```python
#---------------OPTIMAL SOLUTION (NEETCODE)----------------------------

class Solution(object):
    def threeSum(self, nums):
        triplets = []
        nums.sort()

        for i, a in enumerate(nums):
            if i>0 and a==nums[i-1]:
                continue

            l=i+1
            r=len(nums)-1

            while l<r:
                sum = a + nums[l] + nums[r]

                if sum < 0:
                    l+=1
                elif sum > 0:
                    r-=1
                else:
                    triplets.append([a, nums[l], nums[r]]
                    )
                    l+=1
                    while nums[l] == nums[l-1] and l<r:
                        l+=1

        return triplets
```

```python
sola = Solution()
print(sola.threeSum([2,-3,0,-2,-5,-5,-4,1,2,-2,2,0,2,-4,5,5,-10]))
```
